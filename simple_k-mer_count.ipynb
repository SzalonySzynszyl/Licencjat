```python
from Bio import SeqIO
import numpy as np

```

```python

def which_nucleotyde(x): #funkcja przypisująca merom(nukleotydom) w k-merze odpowiednie wartości indeksujące
    l = []
    if x == 'A':
        l.append(0)
    elif x == 'T':
        l.append(1)
    elif x == 'G':
        l.append(2)
    elif x == 'C':
        l.append(3)
    elif x == 'N':#dodałem
        l.append(4)
    else:
        pass
    return l

def k_mers(seq,k=5): #funkcja dzieląca sekwencje na odpowiednie k-mery, które zapisuje w słowniku jako klucze, jako wartości służy częstość występowania poszczególnych k-merów
    lis = []
    d ={}
    for i in range(0,len(seq)-k+1):
        a = seq[i:i+k]
        lis.append(a)
    for i in lis:
        if i in d:
            d[i] += 1
        else:
            d[i] = 1
        
    return d





def position(function): #funkcja, która razem z funkcją which_nucleotyde() tworzy listę opowiednich indeksów dla danej macierzy
    lis2 =[]
    for i in function:
        lis =[]
        for j in i:
            lis.append(which_nucleotyde(j))
        lis2.append(lis)
    return lis2     




def value(function, k): #funkcja służąca do pobierania wartości ze słownika stworzonego przez funkcję k_mers()
    z = []
    for v in function:
        z.append(function[v])
        
    return z[k]

def get_sequence(): #funkcja pobiera z listy wygenerowanej przez SeqIO.parse() same sekwencje, które umieszcza w innej liście
    l = []
    count = 0
    for record in se:
        
        l.append(se[count].seq)
        count -= 1
    return l


names = ('chr1', 'chr2', 'chr3', 'chr4', 'chr5', 'chr6', 'chr7', 'chr8', 'chr9', 'chr10', 'chr11', 'chr12', 'chr13', 'chr14',
         'chr15', 'chr16', 'chr17', 'chr18', 'chr19', 'chr20', 'chr21', 'chr22', 'chr23', 'chr24') #ilość chromosomów w genomie

data = "path"  #ścieżka do pliku

se=list(SeqIO.parse(data, 'fasta')) #zaczytanie pliku fasta przez moduł SeqIO, do listy



a = get_sequence()# zapisanie listy sekwencji do zmiennej

for name in names: #pętla iterująca po poszczególnych chromosomach
    x = -1
    e = k_mers(a[x])# wybieranie i użcie funkcji k_mers() na odpowiedniej sekwencji z listy
    
    n = []
    for p in position(e): #pętla służąca do stworzenia prostej listy z pozycjami, która będzie można przekształcić na tablicę z indeksami
        for z in p:
            for p4 in z:
                n.append(p4)
    
    
        
    tablica = np.zeros((64,5)) #pętla, która ostatecznie tworzy tablicę z poszczególnymi indeksami dla danej sekwencji k-merowej
    k = 192
    j = 5 
    i = 64
    while i > 0:
        j = 5 
        i -= 1
        while j > 0:
            tablica[63-i,5-j] = n[192 - k] 
            j -= 1
            k -= 1
    tablica = tablica.astype('int32')#zmiana typu wartości w tablicy z indeksami, tak aby była możliwa iteracja 
    
    matrix = np.zeros((5,5,5,5,5)) #tworzenie macierzy na wartości frekwencji poszczególnych k_merów(liczba jej wymiarów zależy od dł. k-merów)
    r = 0
    while r <= 63: #pętla, która przy pomocy indeksów z wygenerowanej wcześniej tablicy przypisuje wartości poszczególnych k-merów do wybranych miejsc w macierzy
    
        p1 = 0
        p2 = 0
        p3 = 0
        p4 = 0
        p5 = 0    
        
        p1 += tablica[r,0]
        
        p2 += tablica[r,1]
        
        p3 += tablica[r,2]
    
        p4 += tablica[r,3]
    
        p5 += tablica[r,4]
        
        matrix[p1,p2,p3,p4,p5] = matrix[p1,p2,p3,p4,p5] + value(e,r)
        
        r += 1
    




    np.save("%s.npy" % name, matrix)#zapisanie macierzy z przerobionego chromosomu w pliku binarnym
    a.pop() #skrócenie listy sekwencji a o ostatnią sekwencję i koniec pętli    
    

#plik zapisywany jest z odwrotną nazwą tj. chromosom pierwszy jest zapisany jako ostatni i vice versa, lecz nie ma to znaczenia gdyż podczas analizy i tak wszystkie macierze są ze sobą sumowane
```

```python
def generator(x): #generator ułatwiający zmienianie krotki z imionami dla plików
    z = 1
    b = 'chr'
    while z<x:
        p = b + str(z)
        yield p
        z += 1
a = list(generator(25))
print(a)
```
